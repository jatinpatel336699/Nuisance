```python
from bs4 import BeautifulSoup
from selenium import webdriver 
import numpy as np
import pandas as pd
import datetime
from time import sleep
from tqdm import tqdm

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
sources = ['BOM', 'DEL', 'BLR', 'AMD', 'HYD', 'AMD', 'JLR']
destinations = ['BOM', 'DEL', 'BLR', 'AMD', 'HYD', 'AMD', 'JLR']
```

```python
start = (datetime.date.today())
end = start + datetime.timedelta(days = 2)
start = np.datetime64(start)
end = np.datetime64(end)
print(start)
print(end)
num_days = (end - start).item().days
print(num_days)
print(type(num_days))
```

```python
#url = f'https://www.kayak.co.in/flights/{source}-{destination}/{date}?sort=bestflight_a'

#driver = webdriver.Chrome()

#driver.get(url)

airlines = []
prices = []
duration = []
total_hops = []
date = []
sources_ = []
destinations_ = []
departure_time = []
arrival_time = []
attributes = ['Airline', 'Source', 'Destination', 'Total Stops', 'Duration', 'Date', 'Prices', 'Departure Time', 'Arrival Time']

df = pd.DataFrame(columns = attributes)

#soup = BeautifulSoup(driver.page_source, 'html.parser')

#airplanes_ = soup.find_all('div', class_ = 'nrc6-wrapper')

for i in range(len(sources)):
    for k in range(len(destinations)):
        if (sources[i]==destinations[k]):
            continue

        for j in tqdm(range(num_days)):

            url = f'https://www.kayak.co.in/flights/{sources[i]}-{destinations[k]}/{start+j}?sort=bestflight_a'
            driver = webdriver.Chrome()
            driver.get(url)
            ######################################################################################################
            while(True):
                try:
                    show_more_results = driver.find_element(By.XPATH, "//div[@class = 'ULvh-button show-more-button']")
                    show_more_results.click()
                    driver.implicitly_wait(10)
                    
                except Exception as e:
                    break
            ######################################################################################################    
            soup = BeautifulSoup(driver.page_source, 'html.parser')

            airplanes_ = soup.find_all('div', class_ = 'nrc6-wrapper')
            
            for element in airplanes_:
                sources_.append(sources[i])
                destinations_.append(destinations[k])
                date.append(start+j)
                #print(element.find('div', class_ = 'J0g6-operator-text').text)
                airlines.append(element.find('div', class_ = 'J0g6-operator-text').text)
        
                for element2 in element.find_all('div', class_ = 'xdW8 xdW8-mod-full-airport'):
                    for sub_element_2 in element2.find_all('div', class_ = 'vmXl vmXl-mod-variant-default'):
                        duration.append(sub_element_2.text)
                        #print(sub_element_2.text)

                

                for element3 in element.find_all('div', class_ = 'JWEO'):
                    for sub_element_3 in element3.find_all('span', class_ = 'JWEO-stops-text'):
                        total_hops.append(sub_element_3.text)
                        #print(sub_element_3.text)



                for element4 in element.find_all('div', class_ = 'nrc6-price-section'):
                    for sub_element_4 in element4.find_all('div', class_ = 'f8F1-price-text'):
                        prices.append(sub_element_4.text)
                        #print(sub_element_4.text)

                for element5 in element.find_all('div', class_ = 'vmXl vmXl-mod-variant-large'):
                    timings = [span.text.strip() for span in element5.find_all('span')]
                    print(timings)
                    departure_time.append(timings[0])
                    arrival_time.append(timings[2])
                    
                        
                    
                        

df = pd.DataFrame({
    'Airlines': airlines,
    'Sources': sources_,
    'Destinations': destinations_,
    'Duration': duration,
    'Total Stops': total_hops,
    'Date': date,
    'Departure Time': departure_time,
    'Arrival Time': arrival_time,
    'Prices':prices,
})


    
```
